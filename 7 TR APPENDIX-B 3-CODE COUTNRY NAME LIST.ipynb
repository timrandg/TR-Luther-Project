```python
import pandas as pd
import time

def create_code_country_table():
    country_data = pd.read_csv('country_data.csv')
    country_data['CNT'] = country_data['alpha-3']
    country_data['COUNTRY'] = country_data['name']
    country_data = country_data[['CNT', 'COUNTRY']]
    country_data['COUNTRY'] = country_data.COUNTRY.apply(lambda x : x.upper())
    timestr = time.strftime("%Y%m%d_%H%M%S")
    country_data.to_pickle('./pickles/country_3-code_and_names_' + timestr + '.pkl')
    return country_data
```

```python
c_data = create_code_country_table()
```
